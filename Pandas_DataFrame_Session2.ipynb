```python
#percentage of no of teams won the toss also won the match
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
#percentage of no of times batting 
(ipl[ipl['TossDecision'] == 'bat'].shape[0]/ipl.shape[0])*100
```

```python
#percentage of no of times fielding
(ipl[ipl['TossDecision'] == 'field'].shape[0]/ipl.shape[0])*100
```

```python
#team --->fielding ----> same team --->match won
(ipl[(ipl['TossDecision']=='field') & (ipl['TossWinner'] == ipl['WinningTeam'])].shape[0]/ipl.shape[0])*100
```

```python
#team --->batting ----> same team --->match won
(ipl[(ipl['TossDecision']=='bat') & (ipl['TossWinner'] == ipl['WinningTeam'])].shape[0]/ipl.shape[0])*100
```

```python
#team --->bat ----> same team --->match lose
(ipl[(ipl['TossDecision']=='bat') & (ipl['TossWinner'] != ipl['WinningTeam'])].shape[0]/ipl.shape[0])*100
```

```python
#team --->fielding ----> same team --->match lose
(ipl[(ipl['TossDecision']=='field') & (ipl['TossWinner'] != ipl['WinningTeam'])].shape[0]/ipl.shape[0])*100
```
